This notebook illustrates the bias-variance tradeoff with examples.
The key point is that more complex (many parameters with which to fit to data) estimators do a better job tracking data and its noise but necessarily give different results on different data samples. The better job of tracking the data means lower bias; the differences across datasets means higher variance.

The comparisons in the next cell are based on a function that is a slowly varying sinusoid modulating a more rapidly varying sinusoid.
We generate datasets consisting of the function plus noise.
We compare decision tree and boosting fits to five different datasets from the same distribution,
each with a small number of data samples.
The decision tree depth is kept at 2 to make the decision tree regressors simple (not complex) approximators.
As a result, the decision trees and their avarage cannot track the function very well, particularly on the left half where the slow variation is small.
The trees have a large bias. On the other hand, the different tree fits are all very similar, leading to low variance.

The gradient boosting regressors are more complex approximators. Each individually tracks the function much better than do the decision trees and their average tracks it even better. The bias is smaller than that of the decision trees. On the other hand, the different gradient boosting regressors are all quite different from one another, leading to higher variance.

The reverse holds for the variance: the variance with n_estimators = 256 is larger than the variance with n_estimators = 32.
